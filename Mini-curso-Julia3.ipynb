```julia
# Função Computacional
function meio(x)
    if typeof(x)==Int64 || x-trunc(x)==0
        return(x+.5)
    else
        return((floor(x)+ceil(x))/2) # Se x int, pe, x=10, return 10 (ceil10=10 florr10=10)
    end
end
```

```julia
meio(10.1)
```

```julia
meio(10)
```

```julia
meio(10.0)
```

```julia
function secant(f, a, b)
    m =  (f(b) - f(a)) / (b-a) # slope of secant line
    x -> f(a) + m * (x - a)
end
```

```julia
p(x)=plot!(secant(f,-1+x,0+x),-1+x,0+x)
```

```julia
for i=0:3
    p(i)
end
plot!()
```
